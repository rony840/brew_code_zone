Python Mini Challenge 2 - Simple Calculator

```python
def get_menu_choice(prompt):
    while True:
        try:
            choice = int(input(prompt))
            if choice in (1,2,3,4,5,6):
                return choice
            else:
                print("Please enter a number between 1 to 6.")
        except ValueError:
            print("Invalid input. Please enter a number.")

def get_number_input(prompt):
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Please enter a numeric value.")

def calculate(operation, val1, val2):
    if (operation == 1):
        return val1 + val2
    elif (operation == 2):
        return val1 - val2
    elif (operation == 3):
        return val1 * val2
    elif (operation == 4):
        return val1 / val2

def get_operationSymbol(option):
    if(option==1):
        return " + "
    elif(option==2):
        return " - "
    elif(option==3):
        return " x "
    elif(option==4):
        return " ÷ "

def get_operationString(option):
    if(option==1):
        return "added"
    elif(option==2):
        return "subtracted"
    elif(option==3):
        return "multiplied"
    elif(option==4):
        return "divided by"

def printMenu():
    options = [
        "1. Addition (+)",
        "2. Subtraction (-)",
        "3. Multiplication (x)",
        "4. Division (÷)",
        "5. Clear (AC)",
        "6. Exit"
    ]
    print("Choose an operation from below options:\n")
    for opt in options:
        print(opt)
    return

from IPython.display import clear_output
import time

result = 0
history = ""
while True:
    if(result == 0):
        time.sleep(0.5)
        clear_output(wait=True)
        outputString = 0 if (result == 0 and history=="") else history + " = " + str(result)
        print("Output Display: ",outputString)
        printMenu()
        selectedOption = get_menu_choice("\nEnter a number corresponding to an operation that you want to perform: ")
        if(selectedOption == 6):
            break
        if(selectedOption == 5):
            result = 0
            history = ""
        else:
            num1 = get_number_input("Enter First number: ")
            num2 = get_number_input("Enter Second number: ")
            prevRes = result
            result = calculate(selectedOption,num1,num2)
            history = str(num1) + get_operationSymbol(selectedOption) + str(num2)
    else:
        time.sleep(0.5)
        clear_output(wait=True)
        outputString = history + " = " + str(result)
        print("Output Display: ",outputString)
        printMenu()
        selectedOption = get_menu_choice("\nEnter a number corresponding to an operation that you want to perform: ")
        if(selectedOption == 6):
            break
        if(selectedOption == 5):
            result = 0
            history = ""
        else:
            inputDispString = "Enter next number to be " + get_operationString(selectedOption) + ":"
            num2 = get_number_input(inputDispString)
            prevRes = result
            result = calculate(selectedOption,result,num2)
            history = str(prevRes) + get_operationSymbol(selectedOption) + str(num2)
        
```
