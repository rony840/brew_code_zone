Mini Project 2 - Employee Data Tracker

Requirements:
1. Should print "Employee Data Tracker"
2. Accept employee details via input()
3. Convert salary to numeric using type casting
4. Use operators to calculate yearly salary
5. Display clean output

````python
import re

def get_valid_employee_id():
    pattern = r"^(?=.*\d)[A-Za-z0-9-_]+$"
    while True:
        emp_id = input("Enter Employee ID: ").strip()
        if re.fullmatch(pattern, emp_id) and " " not in emp_id:
            return emp_id
        print("\nError: Employee ID must be alphanumeric (or hyphen/underscore), no spaces, and contain at least one number.")

def get_valid_name():
    pattern = r"^[A-Za-z]+( [A-Za-z]+)*$"
    while True:
        name = input("Enter Employee Name: ").strip()
        if re.fullmatch(pattern, name):
            return name
        print("\nError: Name must contain only letters and single spaces between words.")

def get_valid_salary():
    while True:
        try:
            salary = int(input("Enter Monthly Salary: ").strip())
            if salary > 0:
                return salary
            print("\nError: Salary cannot be zero.")
        except ValueError:
            print("\nError: Salary should be an integer value.")

def get_valid_department():
    pattern = r"^[A-Za-z][A-Za-z0-9\-_ ]*$"
    while True:
        department = input("Enter Department Name: ").strip()
        if re.fullmatch(pattern, department) and not department.isdigit():
            return department
        print("\nError: Department name must start with a letter and be alphanumeric. It cannot be only numbers.")

def employee_data_tracker_main():
    welcome_message = "Employee Data Tracker"
    print(welcome_message)

    emp_id = get_valid_employee_id()
    emp_name = get_valid_name()
    emp_department = get_valid_department()
    monthly_salary = get_valid_salary()
    yearly_salary = monthly_salary * 12

    print("\nEmployee Details:")
    print("ID:", emp_id)
    print("Name:", emp_name)
    print("Department:", emp_department)
    print("Monthly Salary:", monthly_salary)
    print("Yearly Salary:", yearly_salary)

employee_data_tracker_main()
````

In above code we are using regex to add strict validations on user inputs. Regex basically means a regular expression. You can design custom regex patterns based on different needs. How it works? Its simple, if the pattern doesn't match with the string you are trying to compare, it will simply Falsify the comparison based on which you can display an error.

So in below block we are defining a pattern and then inside a while loop we are comparing it with user input and displaying an error if the input string doesnt meet the pattern requirements.
