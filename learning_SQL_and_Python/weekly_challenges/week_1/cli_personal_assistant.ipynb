Common Functions:

```python
from IPython.display import clear_output

def get_menu_choice(options,prompt):
    while True:
        try:
            choice = int(input(prompt))
            if choice in (options):
                return choice
            else:
                options_len = len(options)
                error_msg = "\nPlease enter a number between " + str(options[0]) + " to " + str(options[options_len-1])
                print(error_msg)
        except ValueError:
            print("Invalid input. Please enter a number.")

def print_menu(options, prompt):
    print(prompt)
    for opt in options:
        print(opt)
    return

def get_number_input(prompt):
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Please enter a numeric value.")
```

Temperature Converter Functionality:

```python
def get_unit_symbol(option):
    return "°C" if option == 1 else "°F" if option == 2 else "K"

def get_unit_string(option):
    return "Celsius" if option == 1 else "Farenheit" if option == 2 else "Kelvin"

def convert_temperature(temp, from_unit, to_unit):
    if from_unit == to_unit:
        return temp

    # Celsius
    if from_unit == 1:
        if to_unit == 2:
            return (temp * 9/5) + 32            # Celsius to Farenheit
        else:
            return temp + 273.15                # Celsius to Kelvin

    # Fahrenheit
    if from_unit == 2:
        if to_unit == 1:
            return (temp - 32) * 5/9            # Farenheit to Celsius
        else:
            return (temp - 32) * 5/9 + 273.15   # Farenheit to Kelvin

    # Kelvin
    if to_unit == 1:
        return temp - 273.15                    # Kelvin to Celsius
    else:
        return (temp - 273.15) * 9/5 + 32       # Kelvin to Farenheit

def get_temperature_menu(conversion_unit):
    options = [
        "1. Celsius (°C)",
        "2. Fahrenheit (°F)",
        "3. Kelvin (K)"
    ]
    prompt = "Choose your conversion unit:" if conversion_unit else "Choose your temperature unit:"
    print_menu(options,prompt)
    return

def temperature_main():
    clear_output(wait=False)
    print("Temperature Converter:\n")
    prompt = "Enter Temperature:"
    temp = get_number_input(prompt)
    available_options = [1,2,3]
    get_temperature_menu(False)
    prompt = "Enter a number corresponding to your choice above:"
    unit_option = get_menu_choice(available_options,prompt)
    original_unit = get_unit_symbol(unit_option)
    print("\nOriginal Temperature:",str(temp) + original_unit)
    # ensure different conversion unit
    while True:
        clear_output(wait=False)
        print("Temperature Converter:")
        print(f"\nConvert {temp}{original_unit} to:\n")
        get_temperature_menu(True)
        prompt = "Enter Temperature:"
        conversion_unit_option = get_menu_choice(available_options,prompt)

        if conversion_unit_option == unit_option:
            print("Conversion unit must be different from original unit.")
        else:
            break

    conversion_unit = get_unit_symbol(conversion_unit_option)
    converted_temp = convert_temperature(temp, unit_option, conversion_unit_option)
    conversion_unit_string = get_unit_string(conversion_unit_option)
    clear_output(wait=False)
    print("Temperature Converter:")
    print(f"\n{temp}{original_unit} in {conversion_unit_string} is {converted_temp:.2f}{conversion_unit}")
    return
```

Calculator Functionality:

```python
def calculate(operation, val1, val2):
    if (operation == 1):
        return val1 + val2
    elif (operation == 2):
        return val1 - val2
    elif (operation == 3):
        return val1 * val2
    elif (operation == 4):
        return val1 / val2

def get_operation_symbol(option):
    if(option==1):
        return " + "
    elif(option==2):
        return " - "
    elif(option==3):
        return " x "
    elif(option==4):
        return " ÷ "

def get_operation_string(option):
    if(option==1):
        return "added"
    elif(option==2):
        return "subtracted"
    elif(option==3):
        return "multiplied"
    elif(option==4):
        return "divided by"

def output_display(display):
    print("Calculator:\n")
    print("Output Display:",display)

def get_calculator_menu():
    options = [
        "1. Addition (+)",
        "2. Subtraction (-)",
        "3. Multiplication (x)",
        "4. Division (÷)",
        "5. Clear (AC)",
        "6. Exit"
    ]
    prompt = "Choose an operation from below options:\n"
    print_menu(options,prompt)
    return

def calculator_main():
    clear_output(wait=False)
    result = 0
    history = ""
    while True:
        if(result == 0):
            clear_output(wait=False)
            output_string = 0 if (result == 0 and history=="") else history + " = " + str(result)
            output_display(output_string)
            get_calculator_menu()
            available_options=[1,2,3,4,5,6]
            prompt="\nEnter a number corresponding to an operation that you want to perform: "
            selected_option = get_menu_choice(available_options,prompt)
            if(selected_option == 6):
                return
            if(selected_option == 5):
                result = 0
                history = ""
            else:
                num1 = get_number_input("Enter First number: ")
                clear_output(wait=False)
                output_string = str(num1) + get_operation_symbol(selected_option)
                output_display(output_string)
                num2 = get_number_input("Enter Second number: ")
                previous_result = result
                result = calculate(selected_option,num1,num2)
                history = str(num1) + get_operation_symbol(selected_option) + str(num2)
        else:
            clear_output(wait=False)
            output_string = history + " = " + str(result)
            output_display(output_string)
            get_calculator_menu()
            available_options=[1,2,3,4,5,6]
            prompt="\nEnter a number corresponding to an operation that you want to perform: "
            selected_option = get_menu_choice(available_options,prompt)
            if(selected_option == 6):
                return
            if(selected_option == 5):
                result = 0
                history = ""
            else:
                clear_output(wait=False)
                output_string = str(result) + get_operation_symbol(selected_option)
                output_display(output_string)
                input_display_string = "Enter next number to be " + get_operation_string(selected_option) + ":"
                num2 = get_number_input(input_display_string)
                previous_result = result
                result = calculate(selected_option,result,num2)
                history = str(previous_result) + get_operation_symbol(selected_option) + str(num2)
        
```

```python
def get_assistant_menu():
    options = [
        "1. Convert temperature from one unit to another unit",
        "2. Perform basic calculations",
        "3. Exit"
    ]
    prompt="What can I do for you today?\n"
    print_menu(options,prompt)
    return

def main():
    print("Command-Line Assistant:\n")
    print("Welcome User")
    USER_NAME = input("Please enter your name:")
    print("\nGreetings",USER_NAME,"\n")
    initial_launch = True
    while True:
        if (not initial_launch):
            print("\n\nCommand-Line Assistant:")
            print("\nWelcome back",USER_NAME,"\n")
        get_assistant_menu()
        options = [1,2,3]
        prompt = "Please enter the number corresponding to your choice: "
        selected_options = get_menu_choice(options,prompt)
        if(selected_options == 3):
            clear_output(wait=False)
            print("\nExiting...\n\nSee you later",USER_NAME)
            break
        elif(selected_options == 1):
            temperature_main()
        elif(selected_options == 2):
            calculator_main()
            clear_output(wait=False)
        initial_launch = False

main()
```
