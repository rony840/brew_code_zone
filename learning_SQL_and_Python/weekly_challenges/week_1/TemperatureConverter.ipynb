Python Mini Challenge 1 - Temperature Converter

```python
def get_menu_choice(prompt):
    while True:
        try:
            choice = int(input(prompt))
            if choice in (1, 2, 3):
                return choice
            else:
                print("Please enter 1, 2, or 3.")
        except ValueError:
            print("Invalid input. Please enter a number.")


def get_temperature():
    while True:
        try:
            return float(input("Enter Temperature: "))
        except ValueError:
            print("Temperature must be a numeric value.")


def unit_symbol(option):
    return "°C" if option == 1 else "°F" if option == 2 else "K"

def unit_string(option):
    return "Celsius" if option == 1 else "Farenheit" if option == 2 else "Kelvin"

def convert_temperature(temp, from_unit, to_unit):
    if from_unit == to_unit:
        return temp

    # Celsius
    if from_unit == 1:
        if to_unit == 2:
            return (temp * 9/5) + 32            # Celsius to Farenheit
        else:
            return temp + 273.15                # Celsius to Kelvin

    # Fahrenheit
    if from_unit == 2:
        if to_unit == 1:
            return (temp - 32) * 5/9            # Farenheit to Celsius
        else:
            return (temp - 32) * 5/9 + 273.15   # Farenheit to Kelvin

    # Kelvin
    if to_unit == 1:
        return temp - 273.15                    # Kelvin to Celsius
    else:
        return (temp - 273.15) * 9/5 + 32       # Kelvin to Farenheit


# ================= MAIN PROGRAM =================

options = [
    "1. Celsius (°C)",
    "2. Fahrenheit (°F)",
    "3. Kelvin (K)"
]

print("Choose your temperature unit:")
for opt in options:
    print(opt)

unitOption = int(get_menu_choice("Enter a number corresponding to your choice above: "))
originalUnit = unit_symbol(unitOption)

temp = get_temperature()

# ensure different conversion unit
while True:
    print(f"\nConvert {temp}{originalUnit} to:")
    for opt in options:
        print(opt)

    conversionUnitOption = int(get_menu_choice("Enter a number corresponding to your choice above: "))

    if conversionUnitOption == unitOption:
        print("Conversion unit must be different from original unit.")
    else:
        break

conversionUnit = unit_symbol(conversionUnitOption)

convertedTemp = convert_temperature(temp, unitOption, conversionUnitOption)

conversionUnitString = unit_string(conversionUnitOption)

print(f"\n{temp}{originalUnit} in {conversionUnitString} is {convertedTemp:.2f}{conversionUnit}")
```
