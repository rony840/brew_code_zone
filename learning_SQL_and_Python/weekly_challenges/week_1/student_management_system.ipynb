Mini Project 1 - Student Management System

Requirements:
1. Should print a welcome message "Welcome to Student Management System"
2. Take student details using input()
3. Use variables & data types
4. Type cast age
5. Display formatted student info

````python
import re

def get_valid_student_id():
    pattern = r"^(?=.*\d)[A-Za-z0-9-_]+$"
    while True:
        student_id = input("Enter Student ID: ").strip()
        if re.fullmatch(pattern, student_id) and " " not in student_id:
            return student_id
        print("\nError: Student ID must be alphanumeric (or hyphen/underscore), no spaces, and contain at least one number.")

def get_valid_name():
    pattern = r"^[A-Za-z]+( [A-Za-z]+)*$"
    while True:
        name = input("Enter Student Name: ").strip()
        if re.fullmatch(pattern, name):
            return name
        print("\nError: Name must contain only letters and single spaces between words.")

def get_valid_age():
    while True:
        try:
            age = int(input("Enter Age: ").strip())
            if age > 0:
                return age
            print("\nError: Age cannot be zero.")
        except ValueError:
            print("\nError: Age should be an integer value.")

def get_valid_course():
    pattern = r"^[A-Za-z][A-Za-z0-9\-_ ]*$"
    while True:
        course = input("Enter Course Name: ").strip()
        if re.fullmatch(pattern, course) and not course.isdigit():
            return course
        print("\nError: Course must start with a letter and be alphanumeric. It cannot be only numbers.")

def main():
    welcome_message = "Welcome to Student Management System"
    print(welcome_message)

    student_id = get_valid_student_id()
    student_name = get_valid_name()
    student_age = get_valid_age()
    student_course_enrolled_in = get_valid_course()

    print("\nStudent Details:")
    print("ID:", student_id)
    print("Name:", student_name)
    print("Age:", student_age)
    print("Course:", student_course_enrolled_in)


main()
````

In above code we are using regex to add strict validations on user inputs. Regex basically means a regular expression. You can design custom regex patterns based on different needs. How it works? Its simple, if the pattern doesn't match with the string you are trying to compare, it will simply Falsify the comparison based on which you can display an error.

In ```get_valid_student_id()``` we are using this regex ==> ```r"^(?=.*\d)[A-Za-z0-9-_]+$"``` this ensures that there are no special characters, no spaces and only contains alphanumerics with hyphens and underscores. So in below block we are defining a pattern and then inside a while loop we are comparing it with user input and displaying an error if the input string doesnt meet the pattern requirements. If its a match then we simply return the student_id.

It also puts a strict gaurd that between two words or characters there can only be single space.
