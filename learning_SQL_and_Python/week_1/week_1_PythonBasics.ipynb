```python
# Indentation in python
age = 20

# here is an example of if statement to illustrate indentation
if age >= 18:        # the colon here indicates that
    # next line should have indentation and should only execute
    # when the condition is satisfied
    # this block is indented by 4 spaces
    print("You are old enough to learn python")

# this line is not indented so this line will execute everytime
# regardless of the above if condition outcome
print("Age verification done.")
```

```python
# Variables in Python
age = 25                   # age is an integer
name = "Samantha"          # name is a string
is_employed = False        # is_employed is a boolean
hourly_salary = 21.45      # hourly_salary is a float
```

```python
# Initializing multiple variables with values at once
fruit1, fruit2, fruit3 = "Orange", "Strawberry", "Kiwi"
print(fruit1, fruit2, fruit3)
```

```python
# Assigning single value to multiple variables
x = y = z = 10
print(x,y,z)
"""
Variables should always start with a letter or an underscore.
They are Case-sensitive.
It can contain only alpha-numeric characters & underscores.
Always be descriptive with your variables, so its easier for
everyone to understand the usecase of the variable.
"""
```

```python
# Data Types in Python
# int
age = 25
print(age," => ",type(age))

# str
name = "Samantha"
print(name," => ",type(name))

# bool
is_employed = False
print(is_employed," => ",type(is_employed))

# float
hourly_salary = 21.45
print(hourly_salary," => ",type(hourly_salary))

# complex
x = 1j
print(x, " => ", type(x))

# list
fruits = ["Orange", "Strawberry", "Kiwi"]
print(fruits, " => ", type(fruits))

# tuple
combination = (213, 321, 123, 312, 132, 231)
print(combination, " => ", type(combination))

# range
sequence = range(10)
print(sequence, " => ", type(sequence))

# dict
dictionary = {"name":"Samantha","age":"25"}
print(dictionary, " => ", type(dictionary))

# set
colors = {"red", "crimson", "Ruby"}
print(colors, " => ", type(colors))

# frozenset
fibonacciSeq = frozenset({0,1,1,2,3,5,8,13,21})
print(fibonacciSeq, " => ", type(fibonacciSeq))

# bytes
helloBytes = b"Hello"
print(helloBytes, " => ", type(helloBytes))

# bytearray
arrayBytes = bytearray(10)
print(arrayBytes, " => ", type(arrayBytes))

# memoryview
memoryView = memoryview(bytes(10))
print(memoryView, " => ", type(memoryView))

# NoneType
noneVar = None
print(noneVar, " => ", type(noneVar))
```

```python
# Input & Casting in Python
num1 = input("Enter the first number: ")
num2 = input("Enter the second number: ")
sumOfTwo = num1 + num2
print("Sum of 2 Numbers is wrong here: ", sumOfTwo)
print("Because num1 and num2 is of type:\n", "num1 type: ",type(num1),"num2 type: ",type(num2))
'''
In Python by default all inputs are strings so you need to add type casting if
you want integer or other datatype in input variable
'''
```
