```python
# Indentation in python
age = 20

# here is an example of if statement to illustrate indentation
if age >= 18:        # the colon here indicates that
    # next line should have indentation and should only execute
    # when the condition is satisfied
    # this block is indented by 4 spaces
    print("You are old enough to learn python")

# this line is not indented so this line will execute everytime
# regardless of the above if condition outcome
print("Age verification done.")
```

```python
# Data Types in Python
# int
age = 25
print(age," => ",type(age))

# str
name = "Samantha"
print(name," => ",type(name))

# bool
is_employed = False
print(is_employed," => ",type(is_employed))

# float
hourly_salary = 21.45
print(hourly_salary," => ",type(hourly_salary))

# complex
x = 1j
print(x, " => ", type(x))

# list
fruits = ["Orange", "Strawberry", "Kiwi"]
print(fruits, " => ", type(fruits))

# tuple
combination = (213, 321, 123, 312, 132, 231)
print(combination, " => ", type(combination))

# range
sequence = range(10)
print(sequence, " => ", type(sequence))

# dict
dictionary = {"name":"Samantha","age":"25"}
print(dictionary, " => ", type(dictionary))

# set
colors = {"red", "crimson", "Ruby"}
print(colors, " => ", type(colors))

# frozenset
fibonacciSeq = frozenset({0,1,1,2,3,5,8,13,21})
print(fibonacciSeq, " => ", type(fibonacciSeq))

# bytes
helloBytes = b"Hello"
print(helloBytes, " => ", type(helloBytes))

# bytearray
arrayBytes = bytearray(10)
print(arrayBytes, " => ", type(arrayBytes))

# memoryview
memoryView = memoryview(bytes(10))
print(memoryView, " => ", type(memoryView))

# NoneType
noneVar = None
print(noneVar, " => ", type(noneVar))
```

```python
# String Operations in Python
print("String Operations:\n")
# Concatenation '+': Join two or more strings together
greeting = "Hello"
name = "Jack"
result = greeting + ", " + name     # Concatenates in total 3 strings
print("Concatenation: ",result)

# Repetition '*': Repeats a string for specified number of times
print("Repetition: ","ha" * 3)  # will repeat input string 3 times

# Length 'len(<var_name>/<string>)': returns number of characters in a string
s = "Robert"
print("Length: ",len(s))    # will return length of string

# Membership in/not in: Checks if a substring is present
print("\n\nMembership in/not in strings:\n")
print("Membership in: ","world" in "Hello world!")          # statement is true as "world" exists in "Hello world!"
print("Membership not in: ", "world" not in "Hello world!") # statement is false as "world" does exists in "Hello world!"

# String Conversion 'str(<var_name>/<value:any>)': converts other datatypes
# (like numbers, lists) to string representation
print("\n")
pi = 3.14
text = "The value of pi is " + str(pi)  # Float is converted to string
print("String conversion: ", text)

# Indexing '[<integer_index>]': used for accessing a single character by
# its zero-based index. Negative indices count from the end of the string
print("\n\nIndexing strings:\n")
someString = "Rainbow"
print("Positive Indexing: ", someString[2])     # 2nd index (3rd character) from start of string
print("Negative Indexing: ", someString[-3])    # 3rd character in string from end of string

# Slicing '[<starting_index>:<ending_index>:<step_integer_value>]':
# Extracts substring from the original string using a start:stop:step syntax.
print("\n\nSlicing strings:\n")
largeString = "Hello World! Let's learn Python."
print("Positive start and end: ", largeString[1:5])     # Slice from index 1st to 5th character in string
print("Positive start - no end: ", largeString[19:])    # Discard first 19 characters from start of string
print("Negative start - no end: ", largeString[-19:])   # Include last 19 characters from end of string
print("No start - Positive end: ", largeString[:5])     # Include first 5 characters from start of string
print("No start - Negative end: ", largeString[:-19])   # Discard last 19 characters from end of string
print("Only Step value: ", largeString[::2])            # Skip every other character in string
print("Start - end - step: ", largeString[0:5:2])       # Skip every other character in string


```

```python
# Operators in Python

print("Arithmetic Operators:\n")
# Arithmetic Operators: used for numeric values to perform basic
# mathematical operations
x = 5
y = 2
print("Addition: ", x + y)          # Returns sum of two numbers
print("Subtraction: ", x - y)       # Returns difference between two numbers
print("Multiplication: ", x * y)    # Multiplies one number into another
print("Division: ", x / y)          # Divides one number by another - returns float/decimal value
print("Modulus: ", x % y)           # Divides and Returns the remainder 
print("Exponential: ", x ** y)      # Returns an exponential value
print("Floor division: ", x // y)   # Returns an integer after performing division

print("\n\nAssignment Operators:\n")
# Assignment Operators: used for assigning values to a variable
x = 17; print("Assignment (=): ", x)
x += y; print("Add & Assign (+=): ", x)
x -= y; print("Subtract & Assign (-=): ", x)
x *= y; print("Multiply & Assign (*=): ", x)
x /= y; print("Divide & Assign (/=): ", x)
x %= y; print("Modulo & Assign (%=): ", x)
x = 5; x **= y; print("Exponent & Assign (**=): ", x)
x //= y; print("Floor Divide & Assign (//=): ", x)

print("\n\nComparison Operators:\n")
# Comparison Operators: used for comparing two values
x = 17
y = 12
print("Equal to (==): ", x == y)
print("Not Equal to (!=): ", x != y)
print("Greater than (>): ", x > y)
print("Less than (<): ", x < y)
print("Greater than Equal to (>=): ", x >= y)
print("Less than Equal to (<=): ", x <= y)

print("\n\nLogical Operators:\n")
# Logical Operators: used for combining conditional statements
a = True
b = False
print("Logical AND (and): ", a and b)
print("Logical OR (or): ", a or b)
print("Logical NOT (not): ", not a)

print("\n\nMembership Operators:\n")
# Membership Operators: used to test if a sequence is present in an object
# in - returns True if a sequence with specified value is present in the object
# not in - returns True if a sequence with specified value is not present in the object
stringText = "Python"
numArray = [1, 2, 3, 4]
print("'y' in stringText: ", 'y' in stringText)
print("'z' not in stringText: ", 'z' not in stringText)
print("'2' in numArray: ", 2 in numArray)
print("'5' not in numArray: ", 5 not in numArray)

print("\n\nIdentity Operators:\n")
# Identity Operators: used to compare object, to check if
# they are pointing to same memory location
# is - returns True if both variables are the same object pointing to same memory reference
# is not - returns True if both variables are not the same object pointing to same memory reference
arr1 = [1, 2, 3]
arr2 = [1, 2, 3]
arr3 = arr1
print("arr1 is arr2: ", arr1 is arr2)
print("arr1 is not arr2: ", arr1 is not arr2)
print("arr1 is arr3: ", arr1 is arr3)

print("\n\nBitwise Operators:\n")
# Identity Operators: used to compare object, to check if
# they are pointing to same memory location
# is - returns True if both variables are the same object pointing to same memory reference
# is not - returns True if both variables are not the same object pointing to same memory reference
a = 5
b = 3
print("Bitwise AND (&): ", a & b)   # Sets each bit to 1 if both corresponding bits are 1
print("Bitwise OR (|): ", a | b)    # Sets each bit to 1 if atleast one of the corresponding bit is 1
print("Bitwise XOR (^): ", a ^ b)   # Sets each bit to 1 if only one of the corresponding bit is 1
print("Bitwise NOT (~): ", ~a)      # Inverts all bits 1 to 0 and 0 to 1
print("Left Shift (<<): ", a << 1)  # Shifts bits to left, filling with 0s on right
print("Right Shift (>>): ", a >> 1) # Shifts bits to right, filling with 0s on left
```
